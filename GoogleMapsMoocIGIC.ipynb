```python
import pandas as pd
```

```python
df_librerias=pd.read_csv('dataset.csv')
df_librerias.head()
```

##Creaci√≥n de archivo con extensi√≥n .js

###Qu√© es un archivo geojson? üó∫
Formato para codificar varias estructuras de datos geogr√°ficos basado on JavaScript Object Notation (JSON). Usa una coordenada geogr√°fica del sistema de referencia WGS 1984 y unidades de grados decimales. Soporta geometr√≠as de tipo: Point, LineString, Polygon, MultiPoint, MultiLineString, and MultiPolygon.

```python
tama√±o = len(df_librerias)
i=1
```

```python
jsfile=""
jsfile= "var libreriasData = { 'type':'FeatureCollection',\n"
jsfile = jsfile + "'features':[\n"
for index, row in df_librerias.iterrows():
  nombre=row['Nombre']
  longitud=row['Longitud']
  latitud=row['Latitud']
  if i < tama√±o:
    jsfile = jsfile + " {'type':'Feature', \n 'properties':{'nombre':'"+nombre+"'}, \n 'geometry': {'type':'Point', 'coordinates':["+str(longitud)+","+str(latitud)+"]}},\n"
    i+=1
  else:
    jsfile = jsfile +" {'type':'Feature', \n 'properties':{'nombre':'"+nombre+"'}, \n 'geometry': {'type':'Point', 'coordinates':["+str(longitud)+","+str(latitud)+"]}} \n]};"
```

```python
print(jsfile)
```

###Generar archivo *.js y guardar el c√≥digo en √©l


```python
with open('librerias.js', 'a') as archivo:
  archivo.write(jsfile)
  archivo.close()
```
